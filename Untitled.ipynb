```python
#CREA NUEVA COLUMNA SIN VALORES NULOS Y REPRESENTANDO LAS DIRECCIONES CON NÚMEROS
df['Compañia-Numeric'] = df['COMPAÑÍA'].map({
    'ENTEL' : 1, 
    'CLARO' : 2,
    'WOM' : 3,
    'MOVISTAR' : 4, 
})
df['Compañia-Numeric'].isna().sum()
```

```python
#CREA NUEVA COLUMNA SIN VALORES NULOS Y REPRESENTANDO LAS DIRECCIONES CON NÚMEROS
df['Genero-Numeric'] = df['GÉNERO'].map({
    'Masculino' : 0, 
    'Femenino' : 1, 
})
df['Genero-Numeric'].isna().sum()
```

```python
x1 = pd.DataFrame(df['PAGO_MENSUAL ($)'])
y1 = pd.DataFrame(df['IMPUESTO_ADICIONAL($)'])

x_train = x1[:-20]
x_test = x1[-20:]

y_train = y1[:-20]
y_test = y1[-20:]

modelo = linear_model.LinearRegression()
modelo.fit(x_train, y_train)
y_pred = modelo.predict(x_test)

# The coefficients
print('Coeficientes: \n b1=', modelo.coef_, 'a=', modelo.intercept_)
# The mean squared error
print('Mean squared error (MSE): %.2f'
    % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coeficiente de determinación: %.2f'
    % r2_score(y_test, y_pred))
print('Score del modelo: %.2f' % modelo.score(x_train, y_train))

# Gráficas de la ecuación de regresión
plt.scatter(x_test, y_test, color='black')
plt.plot(x_test, y_pred, color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.title("Ecuación de regresión")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
